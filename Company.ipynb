```python
new_table = df_dividend

new_table = pd.merge(new_table,df_eaning, on='Name', how='outer')
new_table = pd.merge(new_table,df_market, on='Name', how='outer')
new_table = pd.merge(new_table,df_p_e_ratio, on='Name', how='outer',suffixes=('_df1', '_rat'))
new_table = pd.merge(new_table,df_revenue, on='Name', how='outer',suffixes=('_df1', '_rev'))


# Объединение с использованием суффиксов
#result = pd.merge(df1, df2, on='Symbol', how='outer', suffixes=('_df1', '_df2'))

```

Параметр suffixes=('_df1', '_df2') указывает, что к именам дублирующихся столбцов из первого DataFrame будет добавлен суффикс _df1, а к именам дублирующихся столбцов из второго DataFrame — суффикс _df2. Это позволяет избежать конфликтов имен и сохранить все данные.
