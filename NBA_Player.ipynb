```python
#ТОП 10 самых частых колледжей откуда забирают игроков.

df_college = df.groupby('college').count().reset_index()
df_college
```

```python
top_10_college = df_college.sort_values(by='Среднее кол-во очков', ascending=False).head(10)
top_10_college
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(14, 6))

     # Гистограмма роста
plt.subplot(1, 2, 1)
plt.hist(top_10['player_height'], bins=10, color='blue', alpha=0.7)
plt.title('Распределение роста игроков')
plt.xlabel('Рост (см)')
plt.ylabel('Частота')

# Гистограмма веса
plt.subplot(1, 2, 2)
plt.hist(top_10['player_weight'], bins=10, color='orange', alpha=0.7)
plt.title('Распределение веса игроков')
plt.xlabel('Вес (кг)')
plt.ylabel('Частота')

plt.tight_layout()
plt.show()
```

```python
# Условие для категорий по весу
def weight_category(weight):
         if weight < 80:
             return 'Легкий'
         elif 80 <= weight < 100:
             return 'Средний'
         else:
             return 'Тяжелый'

top_10['weight_category'] = top_10['player_weight'].apply(weight_category)

# Подсчет категорий
category_counts = top_10['weight_category'].value_counts()

# Построение круговой диаграммы
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Распределение игроков по весовым категориям')
plt.axis('equal')  # Для круговой диаграммы
plt.show()
```
