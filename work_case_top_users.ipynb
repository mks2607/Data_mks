```python
import pandas as pd
```

```python
data_shpi = pd.read_csv('hpi_period_year.csv')
```

```python
data_shpi['Дата и время приема'] = pd.to_datetime(data_shpi['Дата и время приема'], format='mixed', errors='coerce') # переводим в формат to_datetime
```

```python
#data_shpi['номер недели'] = data_shpi['Дата и время приема'] # добавляю столбец номер недели
#data_shpi['месяц'] = data_shpi['Дата и время приема'] # добавляю столбец месяц
#data_shpi['номер недели'] = pd.to_datetime(data_shpi['номер недели']) # преобразую столбец номер недели в date time
#data_shpi['месяц'] = pd.to_datetime(data_shpi['месяц'])# преобразую столбец номер недели в date time
# как будто код выше лишний для создания столбцов хватит и трех строк код ниже - проверить несколько раз, и сделать рефакторинг

data_shpi['день недели'] = data_shpi['Дата и время приема'].dt.day_name() # добавляю столбец день недели
data_shpi['номер недели'] = data_shpi['Дата и время приема'].dt.isocalendar().week # получаем неделю из столбца дата приема
data_shpi['месяц'] = data_shpi['Дата и время приема'].dt.strftime('%B') # получаем месяц из столбца дата приема
```

```python
data_shpi.tail(2)
```

```python
# тут список ШПИ не пустой - кажется что эта строчка кода и не нужна.
# top_year['ШПИ'] = top_year['ШПИ'].fillna(0)  # Заменяет NaN на 0
```

```python
# Задаем порядок месяцев
mounth_order = ['January', 'July', 'September', 'October', 'August', 'February',
                'March', 'April', 'May', 'June', 'November', 'December']

# Задаем порядок категорий для столбца 'месяц'
data_shpi['день недели'] = pd.Categorical(data_shpi['месяц'], categories=mounth_order, ordered=True)

# Создаем сводную таблицу
top_oprav_year = pd.pivot_table(data_shpi, values='ШПИ', index='ФИО/Название отправителя', 
                                 columns='месяц', aggfunc='count', fill_value=0, observed=False).reset_index()

# Вычисляем итого
top_oprav_year['Итого'] = top_oprav_year.loc[:, top_oprav_year.columns != 'ФИО/Название отправителя'].sum(axis=1)

# Сводная таблица для подсчета по 'Способ доставки'
delivery_counts = data_shpi.groupby(['ФИО/Название отправителя', 'Способ доставки']).size().unstack(fill_value=0)

# Объединяем с основным DataFrame
top_oprav_year = top_oprav_year.merge(delivery_counts, on='ФИО/Название отправителя', how='left')

# Вычисляем доли
top_oprav_year['Доля электронное'] = top_oprav_year['электронное'] / top_oprav_year['Итого']
top_oprav_year['Доля бумажное'] = top_oprav_year['бумажное'] / top_oprav_year['Итого']

# Сортируем по 'Итого'
top_oprav_sort = top_oprav_year.sort_values(by='Итого', ascending=False)

# Выводим результат
top_oprav_sort.head(10)
```

```python
top_oprav_sort.info()
```

```python
data_shpi['день недели'] = pd.Categorical(data_shpi['месяц'], categories=mounth_order, ordered=True)

# Создаем сводную таблицу
top_poluch_year = pd.pivot_table(data_shpi, values='ШПИ', index='Получатель', 
                                 columns='месяц', aggfunc='count', fill_value=0, observed=False).reset_index()

# Вычисляем итого
top_poluch_year['Итого'] = top_poluch_year.loc[:, top_poluch_year.columns != 'Получатель'].sum(axis=1)

# Сводная таблица для подсчета по 'Способ доставки'
delivery_counts = data_shpi.groupby(['Получатель', 'Способ доставки']).size().unstack(fill_value=0)

# Объединяем с основным DataFrame
top_poluch_year = top_poluch_year.merge(delivery_counts, on='Получатель', how='left')

# Вычисляем доли
top_poluch_year['Доля электронное'] = top_poluch_year['электронное'] / top_poluch_year['Итого']
top_poluch_year['Доля бумажное'] = top_poluch_year['бумажное'] / top_poluch_year['Итого']

# Сортируем по 'Итого'
top_poluch_sort = top_poluch_year.sort_values(by='Итого', ascending=False)

# Выводим результат
top_poluch_sort.head(10)
```

```python
# Запись данных в файл Excel - при необходимости, для того что бы загрузить в Data Lens
with pd.ExcelWriter('top_otpravok_2025.xlsx') as writer:
    top_poluch_sort[:10000].to_excel(writer, sheet_name='ТОП получателей', index=False)  # записываем ТОП 10_000 получателей
    top_oprav_sort[:10000].to_excel(writer, sheet_name='ТОП отправителей', index=False)    # записываем ТОП 10_000 отправителей

print("Данные успешно сохранены в файл top_otpravok_2025.xlsx на разных листах.")
```
