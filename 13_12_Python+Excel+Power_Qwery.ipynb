```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_olimpics = pd.read_csv('/content/dataset_olympics.csv') #Импорт данных
df_region = pd.read_csv('/content/noc_region.csv') #Импорт данных
```

```python
df_olimpics.head(5)
```

```python
pivot_olimpics = pd.pivot_table(df_olimpics, values='ID', index='Name', columns='Age', aggfunc='sum', fill_value=0)
pivot_olimpics
```

```python
pivot_olimpics.to_csv('pivot_olimpics.csv',index=False)
```
