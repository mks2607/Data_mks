```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle as pk
```

```python
file_path = '/content/Amazon_products.pkl'

# Открываем .pkl файл для чтения
with open(file_path, 'rb') as file:
    data = pickle.load(file)
```

```python
data.head(5)
```

```python
data.info()
```

```python
#Вывести бестселлеры ТОП 30 с самой дорогой ценой

#top_best = data['is_bestseller']['True']
#top_best = data[data['is_bestseller'] == True]

# . заменить на , в столбце price
data['price'] = data['price'].astype(str).str.replace('.', ',', regex=False)

# Удаляем точки, которые используются как разделители тысяч
data['price'] = data['price'].astype(str).str.replace('.', '', regex=False)

# Заменяем запятую на точку для преобразования в float
data['price'] = data['price'].str.replace(',', '.', regex=False)

# Преобразуем в float, заменяя некорректные значения на NaN
data['price'] = pd.to_numeric(data['price'], errors='coerce')

top_best = data[['price','is_bestseller']].sort_values(by='price', ascending = False)
```

```python
top_best = top_best[top_best['is_bestseller'] == True]
top_best_30 = top_best.head(30)
```

```python
top_best_30.to_csv('top_bests_30.csv', index=False)
```
