```python
data = {
    'Age':[25,28,32,42,41,42,50,59,60]
}

df  = pd.DataFrame(data)
df.plot(y='Age', kind='hist') # показывает разпределение определенных значений на графике - в отличии от bar - который показывает значение каждой строки
plt.title('Age distribution')
plt.show()
```

```python
data = {
    'Category' : ['FIRST','FIRST','SECOND','SECOND','THIRD','THIRD','SECOND'],
    'Value': [10,20,25,45,45,50,50]
}

df = pd.DataFrame(data)
df
```

```python
df.boxplot(by='Category', column = 'Value')
plt.title('Value distribution by category')
plt.show()
```

```python
data = {
    'mounth': ['Jan','Fab','Mar','Apr','May'],
    'Week 1 Temp': [6,10,12,14,17],
    'Week 2 Temp': [8,10,15,17,18],
    'Week 3 Temp': [9,11,14,16,20],
    'Week 4 Temp': [10,15,13,17,22]
}

df = pd.DataFrame(data)
df
```

```python
df.set_index('mounth', inplace=True) # сделали индекс месяцем
df
```

```python
sns.heatmap(df, cmap='coolwarm', annot=True)
plt.title('Weekly temperature data')
plt.show()
```
