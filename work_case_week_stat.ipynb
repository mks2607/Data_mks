# Доля Электронной Доставки

```python
# Шаг 1: Создаем сводную таблицу
dolya_edo = hpi_period_year.pivot_table(index='номер недели', columns='Способ доставки', values='ШПИ', aggfunc='count', fill_value=0)#.reset_index()
# Шаг 2: Вычисляем общую сумму по дате
dolya_edo['Итого ЭД'] = dolya_edo.sum(axis=1)

# Шаг 3: Вычисляем доли
dolya_edo = dolya_edo.div(dolya_edo['Итого ЭД'], axis=0) * 100  # Умножаем на 100 для получения процентов

dolya_edo.reset_index(inplace=True)
dolya_edo.to_string(index=False, header=['номер недели', 'Бумага', 'Электрон', 'Итого ЭД'])
dolya_edo
```

```python
# Фильтрация ЭД для заказных писем
hpi_ordered = hpi_period_year.query("Тип == 'заказное'")
dolya_edo_ordered = hpi_ordered.pivot_table(index='номер недели', columns='Способ доставки', values='ШПИ', aggfunc='count', fill_value=0)
dolya_edo_ordered['Итого'] = dolya_edo_ordered.sum(axis=1)
dolya_edo_ordered = dolya_edo_ordered.div(dolya_edo_ordered['Итого'], axis=0) * 100
dolya_edo_ordered.reset_index(inplace=True)
dolya_edo_ordered.to_string(index=False, header=['Неделя', 'Бумага', 'Электрон', 'Итого'])
dolya_edo_ordered
```

```python

# Фильтрация ЭД для простых писем
hpi_simple = hpi_period_year.query("Тип == 'простое'")
dolya_edo_simple = hpi_simple.pivot_table(index='номер недели', columns='Способ доставки', values='ШПИ', aggfunc='count', fill_value=0)
dolya_edo_simple['Итого'] = dolya_edo_simple.sum(axis=1)
dolya_edo_simple = dolya_edo_simple.div(dolya_edo_simple['Итого'], axis=0) * 100
dolya_edo_simple.reset_index(inplace=True)
dolya_edo_simple.to_string(index=False, header=['Неделя', 'Бумага', 'Электрон', 'Итого'])
dolya_edo_simple
```

# Доля Тарифов

Добавить переменную с разбивкой доли - которую занимает каждый тариф

```python
# Предположим, что hpi_period_year уже существует и содержит нужные данные

# Шаг 1: Создаем сводную таблицу
dolya_tarif = hpi_period_year.pivot_table(index='номер недели', columns='Тариф', values='ШПИ', aggfunc='count', fill_value=0)

# Сброс индекса, чтобы 'номер недели' стал обычным столбцом
dolya_tarif.reset_index(inplace=True)

# Шаг 2: Вычисляем общую сумму по строкам
dolya_tarif['Итого'] = dolya_tarif.sum(axis=1)

# Шаг 3: Вычисляем объединенные значения
dolya_tarif['1+'] = dolya_tarif[31.0] + dolya_tarif[72.0]
dolya_tarif['3+'] = dolya_tarif[[34.5, 38.0, 75.5, 79.0]].sum(axis=1)

# Шаг 4: Вычисляем доли в процентах
dolya_tarif['1+'] = dolya_tarif['1+'] / dolya_tarif['Итого'] * 100
dolya_tarif['3+'] = dolya_tarif['3+'] / dolya_tarif['Итого'] * 100

dolya_tarif['1+'] = dolya_tarif['1+'].round(2) # округдение значений
dolya_tarif['3+'] = dolya_tarif['3+'].round(2) # округдение значений
dolya_tarif['Итого'] = dolya_tarif['1+'] + dolya_tarif['3+'].round(2) # округдение значений
```

# ARPU / ARPPU

```python
hpi_period_year.head(2)
```

```python
# 	ARPU (Average Revenue Per User) = Общий доход за период / количество пользователей за период(и отправители и получатели)
# ARPPU (Average Revenue Per Paying User) = Общий доход за период / Количество платящих пользователей за период(только отправители) 
# Группируем по номеру недели и считаем уникальные значения
unique_users_per_week = hpi_period_year.groupby('номер недели').agg({
    'ШПИ': 'count',
    'ФИО/Название отправителя': 'nunique',
    'Получатель': 'nunique',
    'Тариф': 'sum',
}).reset_index()

# Переименовываем колонки для удобства
unique_users_per_week.columns = ['Номер недели','Итого отправлений', 'Уникальных отправителей','Уникальных получателей', 'Выручка',]

unique_users_per_week['Ср.кол писем на 1 отправителя'] = unique_users_per_week['Итого отправлений'] / unique_users_per_week['Уникальных отправителей'] # добавление новой колонки 
unique_users_per_week['ARPPU'] = unique_users_per_week['Выручка'] / unique_users_per_week['Уникальных отправителей']
unique_users_per_week['ARPU'] = unique_users_per_week['Выручка'] / (unique_users_per_week['Уникальных отправителей']+unique_users_per_week['Уникальных получателей'])

unique_users_per_week
```
