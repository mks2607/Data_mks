```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/content/PremierLeagueSeason2024.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df[df['wins']>10]
```

```python
df= df[df['goals_scored']>50]
df
```

```python
#   

df = df.sort_values(by='goals_scored', ascending=False).head(5)
```

```python

#  
plt.bar(df.index, df['goals_scored'], width=0.4, label=' 1', align='center', color='green')
plt.bar(df.index + 0.4, df['wins'], width=0.4, label=' 2', align='center', color='black')

#    
plt.title('  ')
plt.xlabel('')
plt.ylabel('')
plt.xticks(df.index + 1, df.index)

#  
plt.legend()

#  
plt.show()
```

```python
plt.figure(figsize=(15, 5))

#   
bar_width = 0.5
x = range(len(df['team']))

plt.bar(x, df['goals_scored'], width=bar_width, label='Goals Scored', color='fuchsia', align='center')
plt.bar([p + bar_width for p in x], df['goals_conceded'], width=bar_width, label='Goals Conceded', color='black', align='center')

#    
plt.xlabel('Teams')
plt.ylabel('Goals')
plt.title('Goals Scored vs Goals Conceded per Team')
plt.xticks([p + bar_width/2 for p in x], df['team'], rotation=45)
plt.legend()

#  
plt.tight_layout()
plt.show()
```
