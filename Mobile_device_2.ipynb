```python
import pandas as pd

df = pd.read_csv('/content/user_behavior_dataset.csv')
```

```python
df = df[['User ID','Device Model','Age','Gender']]
```

```python
df
```

```python
df_1 = (df.groupby('Device Model').count())
```

```python
result = df_1.groupby('Device Model').agg({
    'Age': ['sum', 'mean'],   # Сумма и среднее значение продаж
    'Gender': ['sum', 'count'] # Сумма и количество товаров
})
```
