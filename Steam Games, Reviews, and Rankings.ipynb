```python
import pandas as pd

df_descriptions = pd.read_csv('/content/games_description.csv') #https://www.kaggle.com/datasets/mohamedtarek01234/steam-games-reviews-and-rankings
df_ranking = pd.read_csv('/content/games_ranking.csv')
df_reviews = pd.read_csv('/content/steam_game_reviews.csv')
```

```python
#объеденяем таблицу 2 и 3
df_reviews
df_ranking

merged_df = pd.merge(df_reviews,df_ranking, on='game_name',how='outer')
```

```python
reviews_and_ranking = merged_df
```

```python
reviews_and_ranking
```

```python
reviews_and_ranking.info()
```

```python
reviews_and_ranking["game_name"].nunique()
```

```python
reviews_and_ranking.groupby("game_name")['hours_played'].sum()
#суммирует, потому что это все объекты, а нам необходимо стелать столбец hours_played - в тип float
```

```python
#необходимо NaN заменить на 0
#astype()
#float(reviews_and_ranking['hours_played'])

reviews_and_ranking['hours_played'] = pd.to_numeric(reviews_and_ranking['hours_played'], errors='coerce')
```

```python
reviews_and_ranking['hours_played']
```

```python
reviews_and_ranking.groupby("game_name")['hours_played'].sum()
```

```python
reviews_and_ranking['recommendation'].nunique()
```

```python
reviews_and_ranking.head(1)
```

```python
pivot_table = pd.pivot_table(reviews_and_ranking,values='hours_played',  # Значение, которое нужно агрегировать
                             index='game_name',        # Индекс по строкам
                             columns='recommendation',   # Индекс по столбцам
                             aggfunc='sum',       # Функция агрегации
                             fill_value=0)
```

```python
pivot_table
```

```python
df_combined = pd.concat([pivot_table, reviews_and_ranking], axis=0)
```
