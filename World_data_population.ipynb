```python
# Сгруппируем количество населения по continent - и сделаем график - где больше всего людей.
df_continent = df.groupby('continent').sum().reset_index()
df_continent
```

```python
df_continent = df_continent.drop(columns=['rank','cca3','country'])
df_continent
```

```python
df_continent = df_continent.drop(columns=['growth rate','world percentage'])
df_continent
```
