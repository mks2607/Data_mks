```python
import pandas as pd
```

```python
df = pd.read_csv('/content/ds_salaries.csv')
```

```python
df['company_size'].unique()
```

```python
df['company_size'] = df['company_size'].astype(object).str.replace('M', 'Маленькая компания', regex=False)
df['company_size'] = df['company_size'].astype(object).str.replace('S', 'Средняя компания', regex=False)
df['company_size'] = df['company_size'].astype(object).str.replace('L', 'Крупная компания', regex=False)
```

```python
df['company_size'].unique()
```

```python
df['experience_level'].unique()
```

```python
df['experience_level'] = df['experience_level'].astype(object).str.replace('SE', 'Senior', regex=False)
df['experience_level'] = df['experience_level'].astype(object).str.replace('MI', 'Midle', regex=False)
df['experience_level'] = df['experience_level'].astype(object).str.replace('EN', 'Junior', regex=False)
df['experience_level'] = df['experience_level'].astype(object).str.replace('EX', 'Senior+', regex=False)
```

```python
top_100 = df.nlargest(100, 'salary_in_usd')
top_100.to_csv('top_100.csv', index=False)

#Вывести ТОП 100 по зарплате, и сделать отделную таблицу по ним.
```

```python
#Вывести ТОП 100 по минимальной зарплате - и отдельную таблицу по ним.

min_100 = df.sort_values(by='salary_in_usd', ascending=True)
min_100.to_csv('min_100.csv', index=False)
```

```python
count_people = pd.pivot_table(df, values='salary_in_usd', index='company_location', columns='work_year', aggfunc='count', fill_value=0)
count_people
```

```python
mean_salary_years = pd.pivot_table(df, values='salary_in_usd', index='company_location', columns='work_year', aggfunc='mean', fill_value=0)
mean_salary_years
```

```python

mean_salary_years.to_csv('mean_salary_years.csv', index=False)
```

```python
experience_level_tad = pd.pivot_table(df, values='salary_in_usd', index='experience_level', columns='work_year', aggfunc='count', fill_value=0)
experience_level_tad
```

```python
remote_ratio_tad = pd.pivot_table(df, values='salary_in_usd', index='experience_level', columns='remote_ratio', aggfunc='count', fill_value=0)
remote_ratio_tad
```
